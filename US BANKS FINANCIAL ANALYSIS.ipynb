```python
import yfinance as yf
import datetime

# Set the start and end dates
start_date = datetime.datetime(2006, 1, 1)
end_date = datetime.datetime(2016, 1, 1)

# Specify the stock symbol for Bank of America (BAC)
stock_symbol1 = 'BAC'
stock_symbol2 = 'C'
stock_symbol3 = 'GS'
stock_symbol4 = 'JPM'
stock_symbol5 = 'MS'
stock_symbol6 = 'WFC'


# Use yfinance to fetch stock data from Yahoo Finance
BAC = yf.download(stock_symbol1, start=start_date, end=end_date)
C = yf.download(stock_symbol2, start=start_date, end=end_date)
GS = yf.download(stock_symbol3, start=start_date, end=end_date)
JPM = yf.download(stock_symbol4, start=start_date, end=end_date)
MS = yf.download(stock_symbol5, start=start_date, end=end_date)
WFC = yf.download(stock_symbol6, start=start_date, end=end_date)
# Display the stock data
#print(BAC_stock_data.head())

```

** Creating a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers=['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** Using pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list.**

```python
bank_stocks= pd.concat([BAC, C, GS, JPM, MS, WFC],keys=tickers,axis=1)
bank_stocks
```

** Setting the column name levels :**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

** Create a distplot using seaborn of the 2015 returns for Morgan Stanley **

```python
#sns.distplot(returns.loc['2015-01-01':'2016-01-01']['MS'],color is green)
ms_returns_2015 = returns[returns['Date'].dt.year == 2015]['MS']

# Create a distribution plot for 'MS' returns in the year 2015
sns.histplot(ms_returns_2015, kde=True, bins=150, color='g', stat='density')
plt.title('Distribution of MS Returns in 2015')
plt.xlabel('Returns')
plt.ylabel('Density')
plt.show()
```
